### Full Transformation Semigroup

```python
import itertools
def FT_semigroup(n):
    # Define the list of elements to combine
    elements = list(range(1, n+1))
    
    # Generate all possible combinations of n elements with repetition
    combinations = itertools.product(elements, repeat=n)

    # Convert the combinations to a list format
    combinations_list = list(combinations)

    # Return the combinations in a list format
    return combinations_list

# Call the FT_semigroup function for n=3, 4, and 5 and print the length of the resulting list of combinations
combinations_n3 = FT_semigroup(5)
print(f"Length of combinations for n=3: {len(combinations_n3)}")
print(combinations_n3)

#combinations_n4 = FT_semigroup(4)
#print(f"Length of combinations for n=4: {len(combinations_n4)}")
#print(combinations_n4)

#combinations_n5 = FT_semigroup(5)
#print(f"Length of combinations for n=5: {len(combinations_n5)}")
#print(combinations_n5)
```

### Order Preserving Full Transformation Semigroup

```python
def check_increasing_or_constant(combination):
    for i in range(1, len(combination)):
        if combination[i] < combination[i-1]:
            return False
    return True

def Order_Preserving_FT(n):
    # Call the FT_semigroup function to get the list of combinations
    combinations = FT_semigroup(n)
    
    # Check if each combination is in increasing order or constant order, and create a list of combinations that pass the condition
    ordered_combinations = [combination for combination in combinations if check_increasing_or_constant(combination)]
    
    # Return the list of ordered combinations
    return ordered_combinations

# Call the increasing_or_constant_FT function for n=3, 4, and 5
final_combinations_n3 = Order_Preserving_FT(3)
print(final_combinations_n3)
print(len(final_combinations_n3))

#final_combinations_n4 = Order_Preserving_FT(4)
#print(final_combinations_n4)

#final_combinations_n5 = Order_Preserving_FT(5)
#print(final_combinations_n5)
```

### Order Decreasing Full Transformation Semigroup

```python
def Order_Decreasing_FT(tuples):
    # Check if all tuples have the same length.
    if not all(len(tuple) == len(tuples[0]) for tuple in tuples):
        return False

    # Check if all elements in each tuple are less than or equal to the corresponding element in (1, 2, 3).
    passing_tuples = []
    for tuple in tuples:
        domain = [number for number in range(1, len(tuple))]
        for i in domain:
            if tuple[i-1] > i:
                break
        else:
            passing_tuples.append(tuple)
    return passing_tuples

# Call the Order_Preserving_Or_Reversing_FT function for n=3, 4, and 5
combinations_n3 = FT_semigroup(3)
final_combinations_n3 = Order_Decreasing_FT(combinations_n3)
print(final_combinations_n3)
print(len(final_combinations_n3))

#final_combinations_n4 = Order_Decreasing_FT(combinations_n4)
#print(final_combinations_n4)

#final_combinations_n5 = Order_Decreasing_FT(combinations_n5)
#print(final_combinations_n5)
```

### Conjugacy Classes 

```python
import networkx as nx

def conjugacy_class(tuples_list):
    # create an empty dictionary to store the classes
    classes = {}

    # loop through each tuple and create a graph
    for i, tpl in enumerate(tuples_list, start=1):
        # create an empty directed graph
        G = nx.DiGraph()

        # add the edges to the graph based on the elements of the tuple
        for j, el in enumerate(tpl, start=1):
            G.add_edge(j, el)

        # check if the graph is isomorphic to any existing graph
        found = False
        for k, v in classes.items():
            if nx.is_isomorphic(G, v[0]):
                v.append(tpl)
                found = True
                break

        # if the graph is not isomorphic to any existing graph, create a new class
        if not found:
            classes[len(classes)+1] = [G, tpl]

    # print the classes and their edges
    for k, v in classes.items():
        print(f"Class {k}: {len(v)-1} Elements")
        print(f"List of Elements : {[t for t in v[1:]]}")
        #print(f"Nodes: {v[0].nodes()}")
        #print(f"Edges: {v[0].edges()}")
        print("")
```
